```python
import csv

def replace_values(csv_file, column_index, old_value, new_value):
    with open(csv_file, 'r', newline='') as file:
        reader = csv.reader(file)
        rows = list(reader)

    for row in rows:
        if row and len(row) > column_index and row[column_index] == old_value:
            row[column_index] = new_value

    with open(csv_file, 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerows(rows)

# Usage example
csv_file_path = 'Dict_EmolexKatalogApi_Terkejut2 - copy.csv'
column_index = 1  # Index of the column to be modified (0-based index)
old_value = '1.0'
new_value = '1'

replace_values(csv_file_path, column_index, old_value, new_value)
# Save the updated DataFrame to a new CSV file
df.to_csv('Dict_EmolexKatalogApi_Terkejut2.csv', index=False)
df
```

```python
import pandas as pd

def change_column_position(csv_file, column_name, new_position):
    df = pd.read_csv(csv_file)
    columns = df.columns.tolist()
    column_index = columns.index(column_name)
    columns.insert(new_position, columns.pop(column_index))
    df = df.reindex(columns=columns)
    df.to_csv(csv_file, index=False)

# Usage example
csv_file_path = '/content/Dict_EmolexKatalogApi_Sedih - Copy.csv'
column_name = 'word'  # Name of the column to be moved
new_position = 1  # New position (0-based index)

change_column_position(csv_file_path, column_name, new_position)
```
